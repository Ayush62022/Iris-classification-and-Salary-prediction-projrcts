Snice the legit transaction is very large so we can not build the model with the fraudlent transaction which has only 492 row so we take sample from legit transaction of length 492 using sample method
Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample=legit.sample(n=492)
```

```python
legit_sample.head()
```

```python
#concatnating two dataframe
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(len(X_train),len(X_test))
```

```python
print(len(Y_train),len(Y_test))
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
tranning_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
 print(tranning_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(testing_data_accuracy)
```

Now from above tranning and testing accuracy  have less defference 
So our model is accurate
